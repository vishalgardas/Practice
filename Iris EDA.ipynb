```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
print (iris.shape)
```

```python
print(iris.columns)
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind='scatter',x='SepalLengthCm', y='SepalWidthCm');
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="Species", size=5)\
    .map(plt.scatter, 'SepalLengthCm','SepalWidthCm')\
    .add_legend();
plt.show();
```

```python
plt.close();
sns.set_style('whitegrid');
sns.pairplot(iris, hue="Species",height=3);
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue='Species', size=5)\
    .map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm')\
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris, hue='Species', size=6)\
    .map(sns.distplot, 'PetalLengthCm')\
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris, hue='Species', size=6)\
    .map(sns.distplot, 'PetalWidthCm')\
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris, hue='Species', size=6)\
    .map(sns.distplot, 'SepalLengthCm')\
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris, hue='Species', size=6)\
    .map(sns.distplot, 'SepalWidthCm')\
    .add_legend();
plt.show()
```

```python
print('Means:')
print(np.mean(iris[1:50]['PetalLengthCm']))
print("Std-dev")
print(np.std(iris[1:50]['PetalLengthCm']))
```

```python
sns.boxplot(x='Species',y='PetalLengthCm', data=iris)
plt.show()
```

```python
sns.violinplot(x='Species', y='PetalLengthCm', data=iris)
plt.show()
```

```python
sns.jointplot(y="PetalWidthCm", x="PetalLengthCm", data=iris, kind="kde")
plt.show()
```
