```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
russia = pd.read_csv('russia.csv')
russia.head()
```

```python
russia.groupby(['year']).agg({'death_rate':'mean'}).plot.bar(figsize=(15,5))
plt.show()
```

```python
russia.groupby(['year']).agg({'birth_rate':'mean'}).plot.bar(figsize=(15,5))
plt.show()
```

```python
russia['rate_diff']=russia['birth_rate']-russia['death_rate']
```

```python
russia.groupby(['year']).agg({'rate_diff':'mean'}).plot.bar(figsize=(15,5))
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(russia.corr(), annot=True)
```
